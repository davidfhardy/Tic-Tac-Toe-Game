```python
# this function determines which player will make the first move

def firstPlayer():
    validInput = False
    
    playerInput = input("You will be going first. Do you want to be X or O? :> ")
    
    while not validInput:
        if playerInput == "X":
            validInput = True
        elif playerInput == "O":
            validInput = True
        else:
            print(f"{playerInput} is not an option. Select X or O")
    
    return playerInput
```

```python
# this function displays the keypad to show positions to place moves

def displayPositionIndices():
    print("Use the keypad below to choose where you want to place your move\n")
    
    num = 7
    
    # the keypad will be contain 11 rows
    for row in range(0, 11):
        # the keypad will contain 11 columns
        if row == 3 or row == 7:
            print("-----------")
        elif row == 1 or row == 5 or row == 9:
            print(f" {num} | {num+1} | {num+2} ")
            num -= 3
        else:
            print("   |   |   ")
    
    print("\nLet's begin the game:\n")
```

```python
# this function determines the move a player wants to make

def makeMove(turn, moves):
    validInput = False
    
    while not validInput:
        playerInput = input(f"Player {turn}'s move:> ")
        print(" ") # just used to create another line
        position = int(playerInput)
        
        if 1 <= position and position <= 9:
            if moves[position-1] == " ":
                moves[position-1] = turn
                validInput = True
            else:
                print(f"A move has already been placed in position {position}")
        else:
            print(f"{position} is not a valid position")
```

```python
# this function displays the tic tac toe board with the moves that the players have made

def displayBoard(moves):
    num = 6
    
    print("\n")
    
    for row in range(0, 11):
        # the keypad will contain 11 columns
        if row == 3 or row == 7:
            print("-----------")
        elif row == 1 or row == 5 or row == 9:
            print(f" {moves[num]} | {moves[num+1]} | {moves[num+2]} ")
            num -= 3
        else:
            print("   |   |   ")
    
    print("\n")
```

```python
# this function checks if someone has won the game

def checkWinner(moves, symbol):
    return ((moves[6] ==  moves[7] ==  moves[8] == symbol) or # across the top
    (moves[3] ==  moves[4] ==  moves[5] == symbol) or # across the middle
    (moves[0] ==  moves[1] ==  moves[2] == symbol) or # across the bottom
    (moves[6] ==  moves[3] ==  moves[0] == symbol) or # down the middle
    (moves[7] ==  moves[4] ==  moves[1] == symbol) or # down the middle
    (moves[8] ==  moves[5] ==  moves[2] == symbol) or # down the right side
    (moves[6] ==  moves[4] ==  moves[2] == symbol) or # diagonal
    (moves[8] ==  moves[4] ==  moves[0] == symbol)) # diagonal
```

```python
# this function checks if a draw has been reached

def checkDraw(moves):
    if not " " in moves:
        return True
    else:
        return False
```

```python
# driver

print("Welcome to Tic Tac Toe in python!\n")

playerx = "X"
playero = "O"

# see which player will be going first
first = firstPlayer()
counter = 0

# the counter below will be used to determine whose turn it is. When the counter is even, it will be X's turn
# so, if player X goes first, the counter should start at 0. Else, it should start at 1
if first == "O":
    counter = 1

# give instructions for how players can place marks on the tic tac toe board
displayPositionIndices()

# now begin the game
playingGame = True
moves = [" " for i in range(0, 9)] # will hold all the moves made in the game
turn = " " # represent's who's player turn it is: X or O
playerWon = False

while playingGame:
    # check who's turn it is
    if counter%2 == 0:
        turn = playerx
    else:
        turn = playero
    
    playerMoves = makeMove(turn, moves)
    displayBoard(moves)
    playerWon = checkWinner(moves, turn)
    draw = checkDraw(moves)
    
    # check if a player won or a draw has been reached
    if playerWon:
        print(f"Player {turn} has won!")
        break
    elif draw:
        print("A draw has been reached. Better luck next time!")
        break
        
    counter += 1
```
